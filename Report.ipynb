## 1. Introduction

### 1.1 Literature review of the compound 

__Applications of the product__

Fatty Acid Ethyl Esters (FAEE), commonly known as biodiesel, are getting more and more attention as an attractive fuel due to the depleting fossil fuels and because of their versatile applications and the promise they hold for a more sustainable energy future [1]. As the world is facing the environmental challenges posed by fossil fuels, FAEE-based biodiesel emerges as a viable alternative offering numerous advantages [1]. The biggest advantage that biodiesel has over gasoline and petroleum diesel is its environmental friendliness [2]. Even though biodiesel burns similar to the petroleum diesel as far as regulated pollutants are concerned, it has better efficiency than gasoline and reduce engine wear and tear, owing to its favourable combustion properties [2]. Usage of biodiesel will allow a balance to be sought between agriculture, economic development, and the environment [2]. The advantages of biodiesel as diesel fuel are liquid nature portability, ready availability, renewability, higher combustion efficiency, lower sulfur and aromatic content, higher cetane number and higher biodegradability [3].  

__Evaluation of market potential__

The global biodiesel market, valued at approximately USD 36.48 billion in 2022, is expected to surge to USD 79.12 billion by 2032, exhibiting an impressive compound annual growth rate (CAGR) of about 8.1% [4]. Biodiesel, serving as an eco-friendly alternative to traditional diesel, is witnessing increased demand in both power generation and automotive sectors, propelling the expansion of this decentralized market [5]. 

__Biosynthetic pathway/genes__

The biosynthetic pathway in E-coli to produce FAEE: fatty acid ethyl estersuses these enzymes PDH: pyruvate dehydrogenase, ACC: acetyl-CoA carboxylase, BCCP: biotin carboxyl carrier protein, BC: biotin carboxylase, CT: carboxyltransferase, PDC: pyruvate decarboxylase, FAS: fatty acid synthase ADH: alcohol dehydrogenase, TE: thioesterase and FadE: acyl-CoA dehydrogenase.  

First glucose is glycolyzed into precursor pyruvate, which is then decarboxylased into acetaldehyde and dehydrogenased into Acetyl-CoA. This is the start of two pathways that are needed to synthesize FAEE. The acetaldehyde is then dehydrogenased into ethanol. Acetyl-CoA is carboxylased into stearoyl-CoA using also BCCP, BC and CT. Using FAS, malonyl-CoA turns into fatty acyl-ACP. Next step is fatty acyl-ACP esterased into free-fatty acid. From that free-fatty acid is turned using FadE into fatty acyl CoA and the last step of synthesis is reaction of ethanol and fatty acyl CoA into fatty acid ethyl esters [6]. 


 

### 1.2 Literature review of the cell factory 

__Cell factory  Introduction to Escherichia coli and its suitability as a cell factory__

E. coli is a well-studied and well-understood microorganism, which makes it the perfect candidate for genetic manipulations. E. coli has been successfully modified genetically to biosynthesize fatty acids leading to the production of more sustainable biofuels [7]. This was succeeded, as E. coli is suitable for large scale fermentations, hence large-scale biodiesel production and advanced genetic engineering tools are established [8]. Most fatty acids generated within a microbial cell are directed towards the synthesis of phospholipids and TAGs, with a portion undergoing degradation via the -oxidation pathway. Consequently, enhancing the production of free fatty acids (FFAs) could be attained by not only upregulating several genes associated with fatty acid biosynthesis but also by eliminating -oxidation [9]. 

 

__General advantages__

One of the primary advantages of using E. coli as a cell factory is its rapid growth rate, which allows for high production rate of the desired product. This is particularly beneficial in industrial applications where large quantities of product are required in a short period of time. Another important advantage of E. coli is the ease of genetic manipulation, which can be combined with the variety of techniques, making it a great platform of producing a wide range of products [10]. Furthermore, E. coli can be grown in simple and inexpensive media, makes it a cost-effective option in industrial production. 


 

__General disadvantages__

Despite its many advantages, there are also challenges associated with using E. coli as a cell factory. One of the main challenges is endotoxin contamination, specifically lipopolysaccharides (LPS). They are the components of the outer membrane of E. coli, which can be released into the product during the production process, especially during the cell lysis. This can be a significant issue since endotoxins can induce strong immune responses in mammals, which can lead to fever, inflammation, and septic shock in severe cases [11]. Additional steps are often required to remove endotoxins from the product, which can increase the complexity and cost of the production process [12]. Another challenge is the complex downstream processing required to purify the desired product from the host cells and growth medium. The processes can be time-consuming, expensive and may lead to product loss during the process.

__Alternative cell factory__

An alternative cell factory employing Saccharomyces cerevisiae, is engineered to express the foreign wax ester synthase (ws2) sourced from Marinobacter hydrocarbonoclasticus, is employed for the production of fatty acid ethyl esters (FAEEs) using ethanol and acyl-coenzyme A (acyl-CoA) as substrates [13]. For many years, the yeast Saccharomyces cerevisiae has played a crucial role in the brewing industry by converting 6-C sugars like glucose, galactose, and mannose into ethanol. However, it lacks the capability to ferment 5-C sugars such as arabinose and xylose, whereas the bacterium Escherichia coli possess the ability to produce ethanol from pentose sugars [14]. S. cerevisiae has proven to be a more suitable choice for production than E. coli due to its capacity for posttranslational modifications [15]. However, the production yields of certain fatty acids have been observed to be higher in E. coli compared to S. cerevisiae, particularly when posttranslational modifications are unnecessary [15]. Both E. coli and S. cerevisiae possess inherent fermentation capabilities that meet the fundamental criteria for cellulosic ethanol production [16]. E. coli's bacterial metabolic pathway, in particular, demonstrates greater efficiency in fermenting ethanol from the consumed sugars when compared to the yeast pathway [16].

## 3. *If Project category II:* Selection and assessment of existing GSM 

__Assessment of chosen GSM__

The chosen genetic system model (GSM) for this project is Escherichia coli K-12 MG1655, specifically, the iML1515 model, which represents a comprehensive reconstruction of E. coli at the genome level [17]. Due to its close approximation to the wild type, iML1515 retains most pathways in their natural positions and provides a fully mapped-out framework for further investigations.Furthermore, the reconstruction served as a knowledge base for an integrated modeling framework that connects systems biology with structural biology, incorporating information from 1515 protein structures.
The development of iML1515 is based on the iJO1366 model, with notable enhancements. It introduces 184 new genes, 196 additional reactions, and updated confidence scores for existing reactions. Integration with external databases, such as KEGG, PDB, and CHEBI, enhances its utility. Moreover, the model is compatible with computational tools like the Cobra toolbox [17].
Unlike earlier E. coli reconstructions, iML1515 allows for a more in-depth characterization of gene-protein-reaction (GPR) relationships by extending them to catalytic domain resolution. In the context of a genome-scale reconstruction, GPRs establish a formal and explicit connection between genotype and phenotype, associating a gene (G) with the protein (P) responsible for catalyzing a reaction (R) within the network [17].
To assess its performance, the iML1515 model was tested for its ability to predict gene essentiality, achieving a success rate of 93.4%. This marks a significant improvement over the previous model, iJO1366, which had a success rate of 89.8% [17,18]. These results suggest that the GSM is capable of providing reliable predictions for use in this project.
Memote for the iML1515 was also run, with overall score of 91%, the lowest score was in the Annotation gene category, which was 63.32% The lowest consistency score was in Unbounded Flux In Default Medium, which was 88.9%. Overall the Annotation Consistency score was 98.4%. The score for Annotation metabolites was 79%. Also the scores for Reactions and SBO annotations were 81% and 91%. 
Memote is considered a valuable tool in the field of systems biology and bioinformatics. Its automated testing capabilities help researchers and modelers ensure the correctness and reliability of their genome-scale metabolic models. The use of Memote in continuous integration workflows enhances the development and maintenance of these models, contributing to the reproducibility and robustness of computational analyses in systems biology. 
In conclusion from the Memote score, the iML1515 should be able to provide reliable predictions for the project.

__4.3 Potential biodiesel production pathways__ 
Two potential pathways have been analysed. 

The first alternative pathway utilizes glycerol as the only carbon source. In this pathway, glycerol uptake is facilitated by the glycerol kinase (GlpK) and glycerol-3-phosphate dehydrogenase (GlpD), which convert glycerol into dihydroxyacetone phosphate (DHAP). DHAP subsequently enters the central metabolic pathway of glycolysis, which eventually produces stearoyl-CoA. Finally, the stearoyl-CoA and ethanol will undergo a reaction catalyzed by an acyltransferase enzyme (AtfA), producing stearoyl-ethyl ester, a component of FAEE biodiesel. This pathway results in a substantially lower yield value of 2.1e-16 mmol/mmol when compared to the yield obtained from glucose. 


The second potential pathway is producing palmitoyl-CoA instead of stearoyl-CoA as fatty acyl CoA before the systhesis of fatty acid ethyl esters. The yield is 0.111 mmol/mmol from this pathway and it demonstrates a slightly higher yield compared to the stearoyl-CoA biosynthetic pathway (0.105 mmol/mmol).

 __4.6 Compute over-expression and down-regulations targets__ 

Flux Scanning based on Enforced Objective Flux (FSEOF), is an algorithm that identifies fluxes within the metabolic network that rise when product formation is mandated. It incorporates these fluxes as an extra constraint and ultimately suggests a set of reactions requiring overexpression of the involved genes. FSEOF is performed in the WS_DGAT reaction, to identify if any knockouts can be performed to improve the yield of biodiesel.<br>

The FSEOF analysis identified 150 reaction/gene targets which are suitable for up- and down-regulations. 


Reactions exhibiting a flux change above 80% were selectively identified to narrow down potential regulatory candidates, focusing on those with a more substantial impact on biomass yield.
